```python
########################### Currency Synonyms
#################################################################################
# Crypto
result = requests.get('https://web-api.coinmarketcap.com/v1/cryptocurrency/map', {
    'listing_status': 'active',
    'limit': 5000,
    'sort': 'cmc_rank',
}).json()['data']

prefixes = ['', '$', '#', '@', 'r/', 'u/']
prefixes_lower = ['$', 'r/', 'u/']
currency_synonyms = {}
for i, item in enumerate(result):
    slug = item['slug'].lower()
    entity = f'${slug.replace(" ", "_").replace("-", "_")}'
    mapping_items = [
        item["symbol"].lower(), slug, item['name'].lower(), item['name'].lower().replace(' ', '')
    ]
    for mapi in mapping_items:
        for prefix in (prefixes if i < 60 else prefixes_lower):
            currency_synonyms[f'{prefix}{mapi}'] = entity

# Fiat
result = requests.get('https://web-api.coinmarketcap.com/v1/fiat/map', {
    'limit': 5000,
    'sort': 'id',
}).json()['data']
for item in result:
    symbol = item['symbol'].lower()
    name = item['name'].lower()
    sign = item['sign'].lower()
    entity = f'${symbol}'
    mapping_items = [symbol, name]

    # print(f'{symbol} - {sign} - {name}')
    for mapi in mapping_items:
        for prefix in prefixes:
            if f'{prefix}{mapi}' not in currency_synonyms:
                currency_synonyms[f'{prefix}{mapi}'] = entity

# Ensure a currency cant be confused with a word
for word in bert_uncased_vocabulary:
    if word in currency_synonyms:
        currency_synonyms.pop(word)
        currency_synonyms.pop(f'#{word}')

write_pickle(currency_synonyms, 'helper_currency_synonyms')
currency_synonyms
```
